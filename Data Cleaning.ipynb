# Bank Marketing Data Cleaning

```python
import pandas as pd
import numpy as np

data = pd.read_csv("DSA Data Set.csv")
```

```python
data['y'] = data.y.map({'no':0,'yes':1}) #dependent variable
```

```python
data['marital'] = data.marital.map({'divorced':0,'unknown':0,'single':0,'married':1}) #couple = 1 | alone = 0
```

```python
data.head()
```

```python
data.loc[data['age'] <= 21, 'age'] = 0 #teens
data.loc[(data['age'] > 21) & (data['age'] <= 30), 'age'] = 1 #young adults
data.loc[(data['age'] > 30) & (data['age'] <= 50), 'age'] = 2 #adults
data.loc[(data['age'] > 50) & (data['age'] <= 65), 'age'] = 3 #old adults
data.loc[data['age'] > 65, 'age'] = 4 #senior citizens
```

```python
data.head()
```

```python
print(data['age'].max())
print(data['age'].min())
```

```python
#WC(admin,mgmt) = 0 | BC(bc, housemaid,services,technicians) = 1 | owners(self,entr) = 2 | no jobs(retired, unemp, unk) = 3 | students = 4
data['job'] = data.job.map({'admin.':0,'management':0,'housemaid':1,'services':1,'blue-collar':1,'technician':1,'self-employed':2,'entrepreneur':2,'retired':3,'unemployed':3,'unknown':3,'student':4})
```

```python
data.head()
```

```python
data['default'] = data.default.map({'no':0,'unknown':0,'yes':1})
```

```python
data['housing'] = data.housing.map({'no':0,'unknown':0,'yes':1})
```

```python
data['loan'] = data.loan.map({'no':0,'unknown':0,'yes':1})
```

```python
data['contact'] = data.contact.map({'telephone':0,'cellular':1})
```

```python
#previously contacted or not | 0 = 0 | rest = 1
```

```python
data.loc[data['previous'] == 0, 'previous'] = 0
```

```python
data.loc[data['previous'] > 0, 'previous'] = 1
```

```python
data.head()
```

```python
print(data['previous'].max())
print(data['previous'].min())
```

```python
data = data.drop(['duration'], axis = 1)
```

```python
data.head()
```

```python
#10 months, 2 months no data, divided into two groups of 5 | apr-aug = 0 | rest = 1
data['month'] = data.month.map({'mar':1,'sep':1,'oct':1,'nov':1,'dec':1,'apr':0,'may':0,'jun':0,'jul':0,'aug':0})
```

```python
data['day_of_week'] = data.day_of_week.map({'mon':0,'tue':1,'wed':2,'thu':3,'fri':4})
```

```python
data = data.drop(['ModelPrediction'], axis = 1)
```

```python
data.head()
```

```python
#Basic(all basics)=0 | School(high,profess, uni) = 1 | No edu(illi, unkno) = 2
data['education'] = data.education.map({'basic.4y':0,'basic.6y':0,'basic.9y':0,'high.school':1,'professional.course':1,'university.degree':1,'unknown':2,'illiterate':2})
```

```python
data.head()
```

```python
data = data.drop(['campaign'], axis = 1) #unknown
```

```python
data.head()
```

```python
data = data.drop(['pdays'], axis = 1)
```

```python
data['poutcome'] = data.poutcome.map({'nonexistent':0,'success':1,'failure':2})
```

```python
data.head()
```

```python
data = data.drop(['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'nr.employed'], axis = 1)
```

```python
data.euribor3m=data.euribor3m.apply( lambda v: ((v-data.euribor3m.min())/(data.euribor3m.max()-data.euribor3m.min())))
```

```python
data.head()
```
